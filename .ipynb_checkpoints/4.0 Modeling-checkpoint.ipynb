```python
from imblearn.pipeline import Pipeline as Pipeline_With_Sampler
from sklearn.preprocessing import StandardScaler

def make_classifier_pipeline(estimator=None, sampler=None, scaler=StandardScaler()):
    pipeline = Pipeline_With_Sampler([        
        ('scaler', scaler),
        ('sampler', sampler),
        ('estimator', estimator)
    ])
    return pipeline
```

```python
# Adapted from https://scikit-learn.org/stable/auto_examples/calibration/plot_calibration_curve.html
def plot_calibration_plots(clf, name, title='Calibration plots (reliability curve)', baseline=None, baseline_name=None):
    # Calibrated with isotonic calibration
    isotonic = CalibratedClassifierCV(clf, cv=2, method='isotonic')
    # Calibrated with sigmoid calibration
    sigmoid = CalibratedClassifierCV(clf, cv=2, method='sigmoid')

    if baseline is None:
        classifiers = [ (clf, name),
                          (isotonic, name + ' + Isotonic'),
                          (sigmoid, name + ' + Sigmoid') ]
    else:
        classifiers = [ (clf, name),
                          (isotonic, name + ' + Isotonic'),
                          (sigmoid, name + ' + Sigmoid'),
                          (baseline, baseline_name) ]
    
    fig, ax = plt.subplots(figsize=(10, 10))
    _ = ax.plot([0, 1], [0, 1], "k:", label="Perfectly calibrated")
    
    for est, n in classifiers:
        est.fit(X_train, y_train)
        y_pred = est.predict(X_test)
        
        if hasattr(est, "predict_proba"):
            prob_pos = est.predict_proba(X_test)[:, 1]
        else:  # use decision function
            prob_pos = est.decision_function(X_test)
            prob_pos = (prob_pos - prob_pos.min()) / (prob_pos.max() - prob_pos.min())
    
        fraction_of_positives, mean_predicted_value = calibration_curve(y_test, prob_pos, n_bins=10)

        log_loss_score = log_loss(y_test, prob_pos)
        brier_score = brier_score_loss(y_test, prob_pos)
        
        label = f"{n} (Log Loss: {log_loss_score:1.3f}, Brier: {brier_score:1.3f})"
        
        _ = ax.plot(mean_predicted_value, fraction_of_positives, "s-", label=label)
    _ = ax.set_ylabel("Fraction of positives")
    _ = ax.set_ylim([-0.05, 1.05])
    _ = ax.legend(loc="upper left")
    if title is not None:
        _ = ax.set_title(title)
```

```python
from sklearn.metrics import log_loss, brier_score_loss
import numpy as np
from time import time
from sklearn.pipeline import Pipeline

# Adapted from code found at https://imbalanced-learn.org/.
# Original page has been updated. The original code can be found using the Wayback Machine at
# https://web.archive.org/web/20200919051559/https://imbalanced-learn.org/stable/auto_examples/applications/plot_impact_imbalanced_classes.html
class classifier_tracker:
    def __init__(self, X_train, y_train, X_test, y_test):
        self.__eval_scores = pd.DataFrame()
        self.__classifiers = {}
        self.__X_train = X_train
        self.__y_train = y_train
        self.__X_test = X_test
        self.__y_test = y_test
    
    def add_classifier(self, clf, desc=None):
        scores = self.get_scores(clf)
    
        clf_name, clf_desc = self.get_identifier(clf, desc)
        
        if self.__eval_scores.index.isin([(clf_name, clf_desc)]).any():
            self.__eval_scores = self.__eval_scores.drop((clf_name, clf_desc))

        clf_score = pd.DataFrame(scores,
                                 index=pd.MultiIndex.from_tuples([(clf_name, clf_desc)])
                                )
    
        self.__eval_scores = self.__eval_scores.append(clf_score)

        self.__classifiers[(clf_name, clf_desc)] = clf
        
        return clf_score.round(decimals=3)
        
    def get_evaluation_table(self):
        return self.__eval_scores.round(decimals=3)
    
    def get_evaluator(self, clf_name, clf_desc):
        return self.__classifiers[(clf_name, clf_desc)]
    
    def get_scores(self, clf):
        with np.errstate(divide='ignore', invalid='ignore'):
            t_0 = time()
            acc = clf.fit(self.__X_train, self.__y_train).score(self.__X_test, self.__y_test)
            t_1 = time()

            y_probs = clf.predict_proba(self.__X_test)
            y_prob = [ p[1] for p in y_probs ]
            
            mean = np.mean(y_prob)
            log_loss_score = log_loss(self.__y_test, y_prob)
            brier_loss_score = brier_score_loss(self.__y_test, y_prob)


        fit_time = t_1 - t_0        
        
        scores = {
            'mean': mean,
            'log_loss': log_loss_score,
            'brier': brier_loss_score,

            'fit_time': fit_time  
        }
        
        return scores
    
    def get_identifier(self, clf, clf_desc):
        clf_name = self.get_identifier_name(clf)
        clf_desc = self.get_identifier_desc(clf, clf_desc)

        return clf_name, clf_desc
    
    def get_identifier_name(self, clf):
        
        # Get name for estimator. Use the built-in name
        # Account for possibility that the classifier is a pipeline
        if isinstance(clf, Pipeline):
            clf_name = clf[-1].__class__.__name__
        else:
            clf_name = clf.__class__.__name__       
            
        return clf_name
        
    def get_identifier_desc(self, clf, clf_desc):
        # Get the description if not included.
        # If it's a pipeline, take the sampler.
        if clf_desc is None:
            if isinstance(clf, Pipeline_With_Sampler):
                sampler = clf.named_steps['sampler']
                if sampler is not None:
                    clf_desc = sampler.__class__.__name__
                else:
                    clf_desc = ''
            else:
                clf_desc = ''
        
        return clf_desc
```

```python
tracker = classifier_tracker(X_train, y_train, X_test, y_test)
```

## Dummy Classifiers

Two dummy classifiers are used. The first uses the class prior to classify all attempts at a probability equal to the proportion of goals in the dataset. This results in classifying all shots as the majority class (which are misses). The other predicts goals at the same rate as the full sample (but assigns a probability of 1 to each prediction).

```python
from sklearn.dummy import DummyClassifier

dpr = DummyClassifier(
    strategy="prior"
)
tracker.add_classifier(dpr, desc=dpr.strategy)
```

```python
dst = DummyClassifier(
    strategy="stratified",
    random_state=42
)
tracker.add_classifier(dst, desc=dst.strategy)
```

## Logistic Regression

As a baseline, we examine logistic regression methods. We look at logistic regression by itself and with various oversampling and undersampling techniques. Plain logistic regression offers an improvement over both of the dummy classifiers. Additionally, it is notable that none of the oversampling or undersampling methods provide noticeable improvements over base logistic regression. This remains true when applying probability calibration methods.  

The base logistic regression provides metrics that are as good as (or nearly as good as) logistic regression augmented with sampling techniques and/or probability calibration. It is not surprising that probability calibration adds little to the base logistic regression method because logistic regression is already calibrated.

```python
from sklearn.linear_model import LogisticRegression

lr = make_classifier_pipeline(
    estimator=LogisticRegression()
)

tracker.add_classifier(lr)
```

```python
plot_calibration_plots(lr, 'Logistic Regression')
```

Random undersampling randomly picks samples from the majority class, equalizing the number of data points in the majority and minority classes. 

```python
from imblearn.under_sampling import RandomUnderSampler

lrus = make_classifier_pipeline(
    estimator=LogisticRegression(),
    sampler=RandomUnderSampler(random_state=42)
)

tracker.add_classifier(lrus)
```

```python
plot_calibration_plots(lrus, 'Logistic Regression with Random Undersampling')
```

Random oversampling randomly repeats data points in the minority class until counts are equalized with the majority class. 

```python
from imblearn.over_sampling import RandomOverSampler

lros = make_classifier_pipeline(
    estimator=LogisticRegression(),
    sampler=RandomOverSampler(random_state=42)
)

tracker.add_classifier(lros)
```

```python
plot_calibration_plots(lros, 'Logistic Regression with Random Oversampling')
```

SMOTE (Synthetic Minority Over-sampling Technique) is an oversampling technique that synthesizes new minority class members. It does this by randomly selecting a minority class point, determining its nearest neighbors, and building a random linear combination of the neighbors to use as a new minority class member.

```python
from imblearn.over_sampling import SMOTE

lrsm = make_classifier_pipeline(
    estimator=LogisticRegression(),
    sampler=SMOTE(random_state=42)
)

tracker.add_classifier(lrsm)
```

```python
plot_calibration_plots(lrsm, 'Logistic Regression with SMOTE Oversampling')
```

ADASYN (adaptive synthetic sampling) modifies SMOTE to prefer minority class members that are difficult to classify.

```python
from imblearn.over_sampling import ADASYN

lrad = make_classifier_pipeline(
    estimator=LogisticRegression(),
    sampler=ADASYN(random_state=42)
)

tracker.add_classifier(lrad)
```

```python
plot_calibration_plots(lrad, 'Logistic Regression with ADASYN Oversampling')
```

Tomek Link sampling is an undersampling method that removes majority class members that are too close to minority members. A more specific definition can be found [here](https://imbalanced-learn.org/stable/under_sampling.html#tomek-links).

```python
from imblearn.under_sampling import TomekLinks

lrtm = make_classifier_pipeline(
    estimator=LogisticRegression(),
    sampler=TomekLinks()
)

tracker.add_classifier(lrtm)
```

```python
plot_calibration_plots(lrtm, 'Logistic Regression with Tomek Links Undersampling')
```

Edited nearest neighbors uses a nearest-neighbors classification to remove majority class members that disagree with too many neighbors. See [here](https://imbalanced-learn.org/stable/under_sampling.html#edited-nearest-neighbors) for more information.

```python
from imblearn.under_sampling import EditedNearestNeighbours 

lrenn = make_classifier_pipeline(
    estimator=LogisticRegression(),
    sampler=EditedNearestNeighbours()
)

tracker.add_classifier(lrenn)
```

```python
plot_calibration_plots(lrenn, 'Logistic Regression with Edited Nearest Neighbors Undersampling')
```

## Random Forest

Based on the logistic regression case, it appears that sampling methods add little to the results. Consequently, we use only base classifiers from this point forward. Some of these classifiers do incorporate sampling within them, however we will not explicitly add sampling to any classifier.

The base random forest classifier is distinctly inferior than the logistic regression in terms of the log loss metric. Probability calibration makes the method comparable to the logistic regression baseline, however the time it takes to train the classifier means that we will not progress with the random forest.

```python
from sklearn.ensemble import RandomForestClassifier

rf = make_classifier_pipeline(
    estimator=RandomForestClassifier(random_state=42)
)

tracker.add_classifier(rf)
```

```python
plot_calibration_plots(rf, 'Random Forest', baseline=lr, baseline_name='Baseline (Logistic Regression)')
```

By setting the `class_weight` parameter to `balanced`, we can weight the minority class members in a manner to make the minority class equalized with the majority. This does not offer any noticeable improvement over the random forest method alone.

```python
rfb = make_classifier_pipeline(
    estimator=RandomForestClassifier(
        random_state=42,
        class_weight='balanced'
    )
)

tracker.add_classifier(rfb, desc='balanced')
```

```python
plot_calibration_plots(rfb, 'Random Forest (Balanced)', baseline=lr, baseline_name='Baseline (Logistic Regression)')
```

The balanced random forest classifier uses random under sampling to build the random forest. Notably, we see an improvement on the baseline by applying probability calibration.

```python
from imblearn.ensemble import BalancedRandomForestClassifier

brf = make_classifier_pipeline(
    estimator=BalancedRandomForestClassifier(random_state=42, n_jobs=-1)
)

tracker.add_classifier(brf)
```

```python
plot_calibration_plots(brf, 'Balanced Random Forest', baseline=lr, \
                       baseline_name='Baseline (Logistic Regression)')
```

## Gradient Boosting

The histogram-based gradient boosting classifier is a modification of gradient boosting which adds a binning step to divide the training data. This provides a much faster performance than traditional gradient boosting. More information can be found [here](https://scikit-learn.org/stable/modules/ensemble.html#histogram-based-gradient-boosting). 

The default implementation of this classifier outperforms every classifier that we have checked to this point. No improvement is seen by adding probability calibration.

```python
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier

hgbc = make_classifier_pipeline(
    estimator=HistGradientBoostingClassifier(random_state=42)
)

tracker.add_classifier(hgbc)
```

```python
plot_calibration_plots(hgbc, 'Histogram-based Gradient Boosting', baseline=lr, \
                       baseline_name='Baseline (Logistic Regression)')
```

# Feature Importance

We use a permutation importance algorithm to determine the relative importance of the features in the final model. We find that the coordinate features are the most important features, with `is_empty_net` easily in third place. Other features showing as important are the shooter's shot-type selection, whether the shot attempt was a rebound and the number of defensive players on the ice. The game situation (time remaining, period, score) seemed to have little impact on the model, nor did the number of attacking players on the ice. Additionally, despite the fact that shots taken by the home team are more likely to be successful, the feature `is_home` had little importance in the model. 

```python
%%time
# Modified from https://scikit-learn.org/stable/auto_examples/ensemble/plot_forest_importances.html
from sklearn.inspection import permutation_importance

feature_names = X_test.columns
result = permutation_importance(best, X_test, y_test, n_repeats=10, scoring='neg_log_loss', random_state=42, n_jobs=-1)
feature_importances = pd.Series(result.importances_mean, index=feature_names)

fig, ax = plt.subplots(figsize=(14,7))
feature_importances.plot.barh(xerr=result.importances_std, ax=ax)
ax.set_title("Feature importances using permutation on full model", fontsize=16)
ax.set_ylabel("Features", fontsize=16)
ax.set_xlabel("Mean accuracy decrease", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
fig.tight_layout()
plt.show()
```

# Prediction Validity

We can use the predicted probabilities to obtain empirical distributions corresponding to the predicted probabilities of shots that actually resulted in goals ($X_{\text{goal}}$) and those that didn't ($X_{\text{no goal}}$).

```python
probs = best.predict_proba(X_test)[:,1]
prob_df = pd.DataFrame({'prob':probs, 'is_goal': y_test}).sort_values('prob')
```

```python
prob_df['is_goal'] = prob_df['is_goal'].astype(bool)
```

```python
import seaborn as sns
ax = sns.violinplot(x='is_goal', y="prob", data=prob_df)
_ = plt.xlabel('Is Goal', fontsize=14)
_ = plt.ylabel('Predicted Probability', fontsize=14)
```

We now look at the actual counts which determine the empirical distributions above. We stack the histogram for no goal on top of the histogram for successful goals.

```python
n_bins = 50
fig, ax = plt.subplots(figsize=(14,7))
_ = sns.histplot(probs, color="gold", kde=False, ax=ax, bins=n_bins, binrange=[0,1], label='No goal')
_ = sns.histplot(probs[y_test == 1], color="maroon", kde=False, ax=ax, bins=n_bins, binrange=[0,1], label='Goal')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Predicted goal probability', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Prediction Histogram', fontsize=16)
plt.legend(loc='upper right')
plt.show()
```

It is difficult to see the lower region of the histogram above, so we truncate at a maximum count of 200.

```python
fig, ax = plt.subplots(figsize=(14,7))
_ = sns.histplot(probs, color="gold", kde=False, ax=ax, bins=n_bins, binrange=[0,1], label='No goal')
_ = sns.histplot(probs[y_test == 1], color="maroon", kde=False, ax=ax, bins=n_bins, binrange=[0,1], label='Goal')
_ = plt.ylim(0,200)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Predicted goal probability', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Prediction Histogram (Truncated)', fontsize=16)
plt.legend(loc='upper right')
plt.show()
```

We close by examining a P-P plot of the two distributions $X_{\text{goal}}$ and $X_{\text{no goal}}$ to determine similarity. As we should expect from the violin plot above, the asymmetric nature of the plot corresponds to higher predicted likelihoods being more likely to actually correspond to goals. This is the desired outcome.

```python
# P-P plot of distributions (no goal vs goal). Dots are placed at each percentile. Stars at every tenth percentile.
X_goal = probs[y_test == 1]
X_no_goal = probs[y_test == 0]

# Divisions into major/minor intervals. Minor intervals are every percentile, giving 101 different endpoints (0-100). 
# Major are every ten, giving 11 endpoints
n_divs_minor = 101
n_divs_major = 11
divs_minor = np.linspace(0,100,n_divs_minor)
divs_major = np.linspace(0,100,n_divs_major)

percs_no_goal_major = np.percentile(X_no_goal, divs_major)
percs_no_goal_minor = np.percentile(X_no_goal, divs_minor)
percs_goal_major = np.percentile(X_goal, divs_major)
percs_goal_minor = np.percentile(X_goal, divs_minor)

# Marker sizes for minor/major
ms_minor = 6
ms_major = 10

fig, ax = plt.subplots(figsize=(10,10))
_ = plt.plot(percs_no_goal_minor, percs_goal_minor, ls="", marker=".", ms=ms_minor, color='y')
_ = plt.plot(percs_no_goal_major, percs_goal_major, ls="", marker="*", ms=ms_major, color='b')
# Draw dotted line for perfect match
x = np.linspace(0,1)
_ = plt.plot(x,x, color="k", ls="--")
_ = plt.xticks(fontsize=14)
_ = plt.yticks(fontsize=14)
_ = plt.xlabel('No Goal', fontsize=14)
_ = plt.ylabel('Goal', fontsize=14)
_ = plt.title('P-P Plot', fontsize=16)
plt.show()
```
